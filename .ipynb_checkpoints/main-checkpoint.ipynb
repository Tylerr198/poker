```python
from skimage.transform import resize

class PlayingCardDataset(Dataset):
    """Playing card dataset."""

    def __init__(self, csv_file, root_dir, transform=None, resize_shape=(224, 224)):
        """
        Arguments:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
            resize_shape (tuple, optional): Desired shape of the image.
        """
        self.playing_cards = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform
        self.resize_shape = resize_shape

    def __len__(self):
        return len(self.playing_cards)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_name = os.path.join(self.root_dir, self.playing_cards.iloc[idx, 0])
        image = io.imread(img_name)
        label = card_dic[self.playing_cards.iloc[idx, 3]]

        # Resize the image
        image = resize(image, self.resize_shape, anti_aliasing=True)

        if self.transform:
            image = self.transform(image)

        return image, label
```
