```python
import os
import torch
import pandas as pd
from skimage import io, transform
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
plt.ion()   # interactive mode

cards = pd.read_csv('dataset/train_labels.csv')
```

```python

n = 0
img_name = cards.iloc[n,0]
bounding_box = cards.iloc[n, 4:]

print('Image name: {}'.format(img_name))
print('Bounding Coordinates: {}'.format(bounding_box))


def show_cards(image, bounding_box):
    plt.imshow(image)
    plt.scatter(bounding_box[0], bounding_box[1])
    plt.scatter(bounding_box[2], bounding_box[3])
    plt.pause(0.001)
    
plt.figure()

show_cards(io.imread(os.path.join('dataset/train', img_name)), bounding_box)


plt.show()
```
